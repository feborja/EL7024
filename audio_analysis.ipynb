```python
import numpy as np
import soundfile as sf
from IPython.display import Audio
import matplotlib.pyplot as plt
from audio.audio_corrupts import *
import os
import torchaudio
```

## Original Audio

https://www.kaggle.com/datasets/chrisfilo/urbansound8k/data

```python
DATA = "D:\\OneDrive\\OneDrive - Universidad de Chile\\Semestre X\\Teoria de la informacion\\proyecto\\UrbanSound8K\\UrbanSound8K\\audio"
```

```python
signal, Fs = sf.read(os.path.join(DATA + '/fold1/203440-3-0-5.wav'))
signal = signal[:,0]
Audio(data=signal, rate=Fs)
```

```python
# ploting signal
fig, (ax) = plt.subplots(figsize=(12,4))
ax.plot(signal)
ax.set_title('Audio')
ax.set_ylabel('Amplitud')
ax.set_xlabel('Sample')
ax.grid(alpha=0.4)
plt.show()
```

```python
# Se grafica espectrograma para validar frecuencia estimada
fig, (ax) = plt.subplots(figsize=(13,3))
Pxx, freqs, bins, im = ax.specgram(signal, NFFT=512, Fs=Fs, noverlap=0)
ax.set_title('Spectrogram')
ax.set_ylabel('Frecuency [Hz]')
ax.set_xlabel('Time [s]')
```

## Gaussian Noise

```python
signal = torchaudio.load(os.path.join(DATA + '/fold1/203440-3-0-5.wav'))
std = torch.max(signal[0])/10
noise_signal = gaussian_audio(signal, std)
# Audio(data=noise_signal, rate=Fs)
```

```python
# ploting noise signal
fig, (ax) = plt.subplots(figsize=(12,4))
ax.plot(noise_signal[0][0])
ax.set_title('Audio')
ax.set_ylabel('Amplitud')
ax.set_xlabel('Sample')
ax.grid(alpha=0.4)
plt.show()
```
