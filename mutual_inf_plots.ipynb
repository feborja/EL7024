## Information Plane

C칩digo que lee las estimaciones de informaci칩n mutua obtenidas en audio e im치genes y genera los gr치ficos del information plane.

```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
nets = ['clean','filter','gaussian','permute','silent']
titles = ['limpio', 'filtrado', 'ruido gaussiano', 'ventanas permutadas', 'ventanas silenciadas']
layers = ['1','2']

for i, net in enumerate(nets):
    for layer in layers:
        EMIS_X, EMIS_Y = np.load(f'audio_mutualinf/inf_params_l{layer}_{net}.npy')
        colors = np.linspace(1, 100, len(EMIS_X))
        
        plt.plot(EMIS_X, EMIS_Y, 'b-', alpha = 0.2)
        plt.scatter(EMIS_X, EMIS_Y, c = colors, alpha=1)
        plt.text(EMIS_X[-1], EMIS_Y[-1], f'Capa {layer}', fontsize = 10, alpha=0.7) 

    plt.xlabel(r'$I(X,T)$')
    plt.ylabel(r'$I(Y,T)$')
    plt.title(f'Information plane audio {titles[i]}')
    plt.grid(alpha=0.3)
    cbar = plt.colorbar()
    cbar.set_label('% de entrenamiento', rotation=270)
    plt.tight_layout()
    plt.show()
```
